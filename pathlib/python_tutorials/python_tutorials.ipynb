# 2 python基本数据类型


python的标准数据类型有六种：

* 1）Number（数字）
* 2）String（字符串）
* 3）List（列表）
* 4）Tuple（元组）
* 5）Dictionary（字典）
* 6）Set（集合）


其中：

* `不可变数据`（3 个）：Number（数字）、String（字符串）、Tuple（元组）；
* `可变数据`（3 个）：List（列表）、Dictionary（字典）、Set（集合）。



[解释一下](https://blog.csdn.net/dan15188387481/article/details/49864613)`不可变数据类型`和`可变数据类型的区别`：

* id(variable)：`查看当前变量的地址值`

```python
a = 'hello'
print(f'a id:{id(a)}')

b = 'hello'
print(f'b id:{id(b)}')

c = 'hello world'
print(f'c id:{id(c)}')
```
输出结果
```python
a id:140088077884472
b id:140088077884472
c id:140087109174192
```


字符串变量a和变量b的值一样，都是指向hello这个`对象`，也就是说hello在内存中占用的内存地址是固定的，无论有多少个变量指向它，最终分配的内存地址都是固定的。只有数据发生变化了，内存地址才发生变化！


```python
a = [1, 2, 3]
print(f'a {a} id:{id(a)}')

b = [1, 2, 3]
print(f'b {b} id:{id(b)}')

b.append(4)

print(f'b {b} id:{id(b)}')
```
输出结果：
```python
a [1, 2, 3] id:140087109245896
b [1, 2, 3] id:140087109244168
b [1, 2, 3, 4] id:140087109244168
```

列表对象a和对象b虽然值一样，但是指向的内存地址不一样，因此就是可变的数据类型。列表内部增删元素并不会改变列表对象的内存地址！


## 2.1 Number（数字）

1、python支持三种数值类型

* `整型`：不带小数点的整数、负数、0,在python3中没有long整型之分;
* `浮点型`：由`整数部分`和`小数部分`组成，可以用科学计数法表示;
* `复数`：由`实数部分`和`虚数部分`组成，可以用`a+bj`或`complex(a,b)`表示，复数的实数a和虚数b都是`浮点型`。

### 2.1.1 python数字类型转换

python数据类型的转换，只要将数据类型最为函数名即可：

* int(x):将x转换为整数类型;
* float(x)：将x转换为浮点类型;
* complex(x)：将x转换为复数类型，实部为x，虚部为0;
* complex(x, y)：将x和y转换为复数类型，实部为x，虚部为y。


### 2.1.2 十进制、二进制、八进制、十六进制数值类型转换

进制简述：

* `十进制`：就是最常见的数值类型，123
* `二进制`：0b1111011，数字0加字母b（binary）开头
* `八进制`：0o173，数字0加字母o（october）开头
* `十六进制`：0x7b，数字0加字母x（hex）开头

1、`十进制`转换为`其他进制`

* `十进制`转`二进制`：`bin(decimal)`返回二进制字符串
* `十进制`转`八进制`：`oct(decimal)`返回八进制字符串
* `十进制`转`十六进制`：`hex(decimal)`返回十六进制字符串


```python
>>> decimal = 1024
>>> binary = bin(decimal)
>>> binary
'0b10000000000'
>>> octal = oct(decimal)
>>> octal
'0o2000'
>>> hexadecimal = hex(decimal)
>>> hexadecimal
'0x400'
>>> 
```


2、`其他进制`转换为`十进制`

* `二进制`转`十进制`：int(binary_str, 2)
* `八进制`转`十进制`：int(octal_str, 2)
* `十六进制`转`十进制`：int(hexadecimal_str, 2)

```python
>>> binary = '0b10000000000'
>>> int(binary, 2)
1024
>>> octal = '0o2000'
>>> int(octal, 8)
1024
>>> hexadecimal = '0x400'
>>> int(hexadecimal, 16)
1024
>>> 

```
