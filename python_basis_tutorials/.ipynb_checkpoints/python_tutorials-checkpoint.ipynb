# 2 python基本数据类型


python的标准数据类型有六种：

* 1）Number（数字）
* 2）String（字符串）
* 3）List（列表）
* 4）Tuple（元组）
* 5）Dictionary（字典）
* 6）Set（集合）


其中：

* `不可变数据`（3 个）：Number（数字）、String（字符串）、Tuple（元组）；
* `可变数据`（3 个）：List（列表）、Dictionary（字典）、Set（集合）。



[解释一下](https://blog.csdn.net/dan15188387481/article/details/49864613)`不可变数据类型`和`可变数据类型的区别`：

* id(variable)：`查看当前变量的地址值`

```python
a = 'hello'
print(f'a id:{id(a)}')

b = 'hello'
print(f'b id:{id(b)}')

c = 'hello world'
print(f'c id:{id(c)}')
```
输出结果
```python
a id:140088077884472
b id:140088077884472
c id:140087109174192
```


字符串变量a和变量b的值一样，都是指向hello这个`对象`，也就是说hello在内存中占用的内存地址是固定的，无论有多少个变量指向它，最终分配的内存地址都是固定的。只有数据发生变化了，内存地址才发生变化！


```python
a = [1, 2, 3]
print(f'a {a} id:{id(a)}')

b = [1, 2, 3]
print(f'b {b} id:{id(b)}')

b.append(4)

print(f'b {b} id:{id(b)}')
```
输出结果：
```python
a [1, 2, 3] id:140087109245896
b [1, 2, 3] id:140087109244168
b [1, 2, 3, 4] id:140087109244168
```

列表对象a和对象b虽然值一样，但是指向的内存地址不一样，因此就是可变的数据类型。列表内部增删元素并不会改变列表对象的内存地址！

### 2.1.1 python数字类型转换

python数据类型的转换，只要将数据类型最为函数名即可：

* int(x):将x转换为整数类型;
* float(x)：将x转换为浮点类型;
* complex(x)：将x转换为复数类型，实部为x，虚部为0;

### 2.1.2 十进制、二进制、八进制、十六进制数值类型转换

进制简述：

* `十进制`：就是最常见的数值类型，123
* `二进制`：0b1111011，数字0加字母b（binary）开头
* `八进制`：0o173，数字0加字母o（october）开头
* `十六进制`：0x7b，数字0加字母x（hex）开头

1、`十进制`转换为`其他进制`

* `十进制`转`二进制`：`bin(decimal)`返回二进制字符串
* `十进制`转`八进制`：`oct(decimal)`返回八进制字符串
* `十进制`转`十六进制`：`hex(decimal)`返回十六进制字符串


```python
>>> decimal = 1024
>>> binary = bin(decimal)
>>> binary
'0b10000000000'
>>> octal = oct(decimal)
>>> octal
'0o2000'
>>> hexadecimal = hex(decimal)
>>> hexadecimal
'0x400'
>>> 
```


2、`其他进制`转换为`十进制`

* `二进制`转`十进制`：int(binary_str, 2)
* `八进制`转`十进制`：int(octal_str, 2)
* `十六进制`转`十进制`：int(hexadecimal_str, 2)

```python
>>> binary = '0b10000000000'
>>> int(binary, 2)
1024
>>> octal = '0o2000'
>>> int(octal, 8)
1024
>>> hexadecimal = '0x400'
>>> int(hexadecimal, 16)
1024
>>> 

```

###2.2.4 字符串格式化1：使用`%`的字符串格式化

Python 支持`格式化字符串`的`输出` 。尽管这样可能会用到非常复杂的表达式，但最基本的用法是将一个值插入到一个有字符串格式符 %s 的字符串中。

在 Python 中，字符串格式化使用与 C 中 sprintf 函数一样的语法。`print() 函数`使用以`%开头`的`转换说明符`对各种类型的数据进行`格式化输出`，具体请看下表：


#### 2.2.4.1 python转换格式符


1、[python转换格式符](http://c.biancheng.net/view/2177.html)


转换格式符 | 说明
:----|:-----
`%d 、%i`  |  转换为带符号的十进制整数
`%b `  |  转换为带符号的二进制整数
`%o `  |  转换为带符号的八进制整数
`%x、%X` |  转换为带符号的十六进制整数
`%s  ` | 使用 str() 函数将表达式转换为字符串
`%r  `|  使用 repr() 函数将表达式转换为字符串
`%c  ` |  单个字符（格式化字符及其 ASCII 码）
`%e  ` |  转化为科学计数法表示的浮点数（e 小写）
`%E  ` |  转化为科学计数法表示的浮点数（E 大写）
`%f、%F ` |  转化为十进制浮点数
`%g `  |  自动选择使用 %f 或 %e 格式
`%G `  |  自动选择使用 %F 或 %E 格式
`%% ` |   字符"%"，显示百分号%



`转换格式符`（Conversion Specifier）只是一个占位符，它会被后面表达式（变量、常量、数字、字符串、加减乘除等各种形式）的值代替。


2、实例：

```python
>>> age = 32
>>> name = "Tom"
>>> print("%s age is %d" % (name, age))
Tom age is 32
>>> print("%s age is %i" % (name, age))
Tom age is 32
>>> print("%s age is %o" % (name, age))
Tom age is 40
>>> print("%s age is %x" % (name, age))
Tom age is 20
>>> print("%s age is %X" % (name, age))
Tom age is 20
>>> print("%s age is %f" % (name, age))
Tom age is 32.000000
>>> print("%s age is %F" % (name, age))
Tom age is 32.000000
>>> print("%s age is %.2f" % (name, age))
Tom age is 32.00
>>> print("%s age is %r" % (name, age))
Tom age is 32
>>> 
```


#### 2.2.4.2 %格式化：指定最小输出宽度


当使用上表中的转换说明符时，可以使用下面的格式指定最小输出宽度（至少占用多少个字符的位置）：

* `%10d`： 表示输出的整数宽度至少为 10
*`%20s`： 表示输出的字符串宽度至少为 20



```python
>>> num = 123456789
>>> print("show num: %06d" % num)
show num: 123456789
>>> print("show num: %09d" % num)
show num: 123456789
>>> print("show num: %9d" % num)
show num: 123456789
>>> print("show num: %12d" % num)
show num:    123456789
>>> 
```

`注意：`

* 当%n中的n值小于等于后面显示数字字符长度时，完整显示数字
* 当%n中的n值大于后面数字字符长度时，就在`数字左侧`填补`n-len(str(num))`个空格


#### 2.2.4.3 %格式化：指定对其方式


1、`-`：左对齐，在左边填充空格

```python
>>> url = "www.baidu.com"
>>> print("url: %20s" %url)
url:        www.baidu.com
>>> print("url: %-20s" %url)
url: www.baidu.com       
>>> 
```

2、`+`：右对齐（+加号可以省略），在右边填充空格


#### 2.2.4.4 %格式化：指定小数精度


* `%m.nf`
* `.nf`

```python
f = 3.141592653
# 最小宽度为8，小数点后保留3位
print("%8.3f" % f)
# 最小宽度为8，小数点后保留3位，左边补0
print("%08.3f" % f)
# 最小宽度为8，小数点后保留3位，左边补0，带符号
print("%+08.3f" % f)
```

输出结果：

```python
   3.142
0003.142
+003.142
```

### 2.2.9 字符串的内建函数

注意：

以下所有的`S`都是值指代一个字符串对象S，不要误理解就是一个字母`S`


1、`S.capitalize()`：将字符串S的`首字母大写`

```python
>>> a = 'hello'

>>> a
```

2、`S.casefold()`：casefole()函数是python3.3之后版本引入的，其作用和`lower()`方法相似，都是将`字符串中所有的大写字母都转换为小写`


二者的区别：
* lower()方法只对`ASCII`编码有效，也即是对`A-Z`有效
* casefold()方法可以把`非汉语或英文`中的`大写转换为小写`

```python
>>> S1 = "Runoob EXAMPLE....WOW!!!" #英文
>>> S2 = "ß"  #德语
>>> print( S1.lower() )
runoob example....wow!!!
>>> print( S1.casefold() )
runoob example....wow!!!
>>> print( S2.lower() )
ß
>>> print( S2.casefold() ) #德语的"ß"正确的小写是"ss"
ss
>>> 
```

3、`S.center(width, fillchar)`：把字符串S填充到fillchar字符中间

* `width`:是填充字符和S字符串的总长度
* `fillchar`：是要在S字符串两边填充的字符


* sub：搜索的子字符串
* start：字符串开始搜索的位置。默认为第一个字符,第一个字符索引值为0。
* end：字符串中结束搜索的位置。字符中第一个字符的索引为 0。默认为字符串的最后一个位置。

```python
>>> str = "www.google.com"
>>> str.count("o")
3
>>> str.count("oo")
1
>>> str.count("ww")
1
>>> str.count("o", 10, 13)
1
```

* `encoding`：是要使用的`编码方式`，如：`UTF-8`
* `errors`：置不同错误的处理方案。默认为 'strict',意为编码错误引起一个UnicodeError。 其他可能得值有 'ignore', 'replace', 'xmlcharrefreplace', 'backslashreplace' 以及通过 codecs.register_error() 注册的任何值。


可以使用：`chardet`查看字符的编码方式（在python3中已经不支持`S.decode()方法`）：

`注意：`

1、递归函数定义

`递归函数`：如果一个函数在`内部调用自己`，这个函数就是`递归函数`。


2、简单递归函数实例

计算阶乘n! = 1 x 2 x 3 x ... x n，用函数fact(n)表示，可以看出：

$fact(n)=n!=1\times2\times3\times\cdot\cdot\cdot\times(n-1)\times n=(n-1)!\times n=fact(n-1)\times nfact(n)=n!=1×2×3×⋅⋅⋅×(n−1)×n=(n−1)!×n=fact(n−1)×n$

所以，`fact(n)`可以表示为`n x fact(n-1)`，只有n=1时需要特殊处理，以此类推`fact(n-1)`可以表示为`(n-1)xfact(n-2)`...

于是，fact(n)用递归的方式写出来就是：

```python
def fact(n):
    if n==1:
        return 1
    return n * fact(n - 1)

```

然后测试一下递归函数：

```python
>>> fact(5)

>>> fact(100)

```

### 3.2.3 生成器（generator）

例如我们通过`列表生成式`创建一个列表，但是受`内存的限制`，列表的容量肯定是有限的，例如我们创建一个`100万`个元素的列表，而如果我们只用到前几个元素，那后面绝大多元素占用的空间都白白浪费了！


如果`列表中的元素`可以`按照某种算法推算`出来，那么在`循环`的过程中就可以`推算`出后面的元素，而不必创建完整的list，从而`节省大量的空间`，因此`生成器`应运而生！

#### 3.2.3.1 生成器（generator）

1、`生成器`（`generator`）的定义

>python中，通过`一边循环一边推算后面的元素`，称为`生成器（generator）`


2、创建生成器：把`列表生成式`的`[]`改为`()`即可创建一个生成器

```python
>>> L
[1, 4, 9, 16, 25, 36, 49, 64, 81, 100]
>>> g = (x * x for x in range(1, 11))
>>> g
<generator object <genexpr> at 0x7f195443df90>
>>> 
```

3、如可打印出生成器中的元素

通过`next(generator)`获得generator的`下一个返回值`

```python
>>> g = (x * x for x in range(1, 11))
>>> g
<generator object <genexpr> at 0x7f195443df90>
>>> next(g)
1
>>> next(g)
4
>>> next(g)
9
>>> next(g)
16
>>> next(g)
25
>>> next(g)
36
>>> next(g)
49
>>> next(g)
64
>>> next(g)
81
>>> next(g)
100
>>> next(g)
Traceback (most recent call last):
  File "<stdin>", line 1, in <module>
StopIteration
>>> 
```


`注意：`

>当next获取到`最后一个元素`，直到没有更多元素时，会抛出`StopIteration`错误！因此正确的方法是通过for循环，而且不会报异常错误，因为`生成器（generator）`也是`可迭代（iterable）`对象。


4、使用for循环获取生成器中的元素

```python
>>> g = (x * x for x in range(1, 11))
>>> g
```

* `yield`：的作用有点类似`return`，当遇`yield`就会先返回`yield后的值`，就相当于是`调用了一次next()`，仔细琢磨一下是不是这样的！

#7 进程和线程 

##7.1 进程和线程概念介绍


现在的操作系统比如`Mac OS X`，`UNIX`，`Linux`，`Windows`等，都是支持`“多任务”`的操作系统。

1、什么叫多任务？

>简单来说就是操作系统可以`同时运行多个任务`。例如，你在写使用电脑在word上写论文，一边写，一边在Chrome浏览器查阅相关的资料，还一边使用网易云听音乐，此时就至少有三个任务正在运行。还有很多的任务在后台执行着，只是没有在桌面显示而已。


现在的多核CPU已经非常普遍了，即使是过去的单核CPU也是可以执行多任务！但是由于`CPU执行代码都是顺序执行的`，那么`单核CPU是怎么执行多任务的呢`？

2、`单核CPU`是如何`执行多任务`的？

>由于`CPU执行任务`都是`顺序执行的`，因此`单核CPU`如果要实现多任务同时执行，就只能通过`交替执行多个任务`，例如先执行A任务0.01秒，切换到任务B,执行B任务0.01秒，再切换到任务C，执行0.01秒...，这样反复执行下去，由于CPU执行的速度很快，因此我们感觉不到任务是在交替执行的，因此直觉上也就给我一种所有任务都是同时执行的！


3、什么是`进程`（`Process`）？

> 对于操作系统而言，`一个任务就是一个进程（Process）`

例如：

* 打开一个浏览器就是`启动一个浏览器进程`
* 打开一个记事本就是`启动一个记事本进程`
* 打开两个记事本就是`启动两个记事本进程`


4、什么是`线程`（`Thread`）

>有些`进程`还`不止同时干一件事`，比如Word，它可以`同时`进行`打字`、`拼写检查`、`打印`等事情。在`一个进程内部`，要`同时干多件事`，就需要`同时`运行多个`“子任务”`，我们把`进程内`的这些`“子任务”`称`为线程`（`Thread`）。


* `一个进程`至少要执行一个任务
* `一个进程`至少要有一个`线程`
* 像word这种`复杂的进程`，可以有`多个线程同时执行`，`多线程`的执行方式和多进程是一样的，也是由`操作系统在多个线程之间快速切换实现`，让每个线程都短暂的交替运行，看起来像是同时执行一样。
* 真正的`同时执行多个线程`需要`多核CPU`才可以实现。例如两个线程分别放在CPU的两个内核上同时执行，此时就没有交替实现的同时执行多个线程。


5、python中如何实现多任务同时执行？


* 1）`多进程模式`：启动`多个进程`，然后每个进程`只有一个线程`，此时多个进程可以一块执行多个任务。
* 2）`多线程模式`：启动`一个进程`，然后在这个进程中启动`多个线程`，多个线程也可以一块执行多个任务。
* 3）`多进程模式`+`多线程模式`：，就是启动多个进程，每个进程再启动多个线程，这样同时执行的任务就更多了，当然这种模型更复杂，实际很少采用




同时执行多个任务通常各个任务之间并不是没有关联的，而是需要`相互通信`和`协调`，有时，任务1必须暂停等待任务2完成后才能继续执行，有时，任务3和任务4又不能同时执行，所以，多进程和多线程的程序的复杂度要远远高于我们前面写的`单进程`、`单线程`的`程序`。

因为复杂度高，调试困难，所以，不是迫不得已，我们也不想编写多任务。但是，有很多时候，没有多任务还真不行。想想在电脑上看电影，就必须由一个线程播放视频，另一个线程播放音频，否则，单线程实现的话就只能先把视频播放完再播放音频，或者先把音频播放完再播放视频，这显然是不行的。


6、python既支持`多进程`，也支持`多线程`！

7、小结

* `线程`是执行的单元，而进程中至少有一个线程组成
* 如何调度`进程和线程`，完全由操作系统决定，程序不能决定什么时候执行，执行多长时间
* `多进程`和`多线程`的程序涉及到`同步`、`数据共享`的问题，编写起来更复杂！



##7.2 多进程（multiprocessing）

python是跨平台的，提供了跨平台的多进程支持，`multiprocessing`是跨平台的`多进程模块`


###7.2.1 创建子进程

`multiprocessing`模块提供了一个`Process类`来代表一个`进程对象`，下面的例子演示了`启动一个子进程`并等待其结束：


```python
from multiprocessing import Process
import os

# 子进程要执行的代码
def run_proc(name):
    print(f"Run child process {name} ({os.getpid()})")


if __name__ == "__main__":
    # 父进程
    print(f"Parent process {os.getpid()}")
    # 创建一个子进程对象
    p = Process(target=run_proc, args=("test", ))
    print(f"Child process will start")
    # p.start()启动子进程
    p.start()
    # p.join()等待子进程执行完毕再继续往下运行
    p.join()
    print("Child process end")
```

输出结果：

```python
Parent process 7486
Child process will start
Run child process test (14118)
Child process end
```


1、创建一个子进程Process对象

>`p = Process(target=run_proc, args=("test", ))`

Process需要两个参数：
* target要执行的函数对象名
* args函数的参数，如果只有一个参数，也要以元素的形式传递，记住一个参数时`必须加上逗号`

2、调用start()方法，启动子进程

>`p.start()`

3、调用join()方法，等待子进程执行结束后再继续往下运行，通常用于`子进程间的同步`

>`p.start()`


###7.2.2 创建多个子进程，使用进程池Pool

如果要`启动大量`的`子进程`，可以用`进程池`的方式`批量创建子进程`：

```python
from multiprocessing import Pool
import os, time, random

def long_time_task(name):
    print(f"Run task {name} {(os.getpid())}")
    start = time.time()
    # random.random() 随机生成一个[0,1)之间的数
    time.sleep(random.random()*3)
    end = time.time()
    print(f"Task {name} runs {(end-start):0.2f} seconds")
    

if __name__ == "__main__":
    print(f"Parent process {os.getpid()}")
    # 创建7个子进程 
    p = Pool(6)
    for i in range(7):
        p.apply_async(long_time_task, args=(i, ))
    print("Waiting for all subprocess done ...")
    p.close()
    p.join()
    print("All subprocesses done.")
```

输出结果：

```python
Parent process 7486
Run task 0 3838
Run task 1 3839
Run task 4 3842
Run task 3 3841
Run task 2 3840
Run task 5 3843
Waiting for all subprocess done ...
Task 2 runs 0.18 seconds
Run task 6 3840
Task 6 runs 0.14 seconds
Task 0 runs 0.39 seconds
Task 3 runs 0.48 seconds
Task 5 runs 1.21 seconds
Task 1 runs 2.21 seconds
Task 4 runs 2.64 seconds
All subprocesses done.
```


对`Pool对象`调用`join()`方法会`等待`所有子进程执行完毕，调用`join()`之前必须先调用`close()`，`调用close()之后`就不能继续添加新的`Process`了。


`注意：`

输出的结果，task `0`、`1`、`2`、`3`、`4`、`5`这6个子进程是立刻执行的，而task `6` 这个子进程要等待前面某个task完成后才执行，这是因为`Pool`的默认大小是`CPU的核数`，如果你的`CPU`有`n个核数`，那你至少要创建`n+1个子进程`才能看到上面的等待效果。
