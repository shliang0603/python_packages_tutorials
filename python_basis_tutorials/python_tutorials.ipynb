# 2 python基本数据类型


python的标准数据类型有六种：

* 1）Number（数字）
* 2）String（字符串）
* 3）List（列表）
* 4）Tuple（元组）
* 5）Dictionary（字典）
* 6）Set（集合）


其中：

* `不可变数据`（3 个）：Number（数字）、String（字符串）、Tuple（元组）；
* `可变数据`（3 个）：List（列表）、Dictionary（字典）、Set（集合）。



[解释一下](https://blog.csdn.net/dan15188387481/article/details/49864613)`不可变数据类型`和`可变数据类型的区别`：

* id(variable)：`查看当前变量的地址值`

```python
a = 'hello'
print(f'a id:{id(a)}')

b = 'hello'
print(f'b id:{id(b)}')

c = 'hello world'
print(f'c id:{id(c)}')
```
输出结果
```python
a id:140088077884472
b id:140088077884472
c id:140087109174192
```


字符串变量a和变量b的值一样，都是指向hello这个`对象`，也就是说hello在内存中占用的内存地址是固定的，无论有多少个变量指向它，最终分配的内存地址都是固定的。只有数据发生变化了，内存地址才发生变化！


```python
a = [1, 2, 3]
print(f'a {a} id:{id(a)}')

b = [1, 2, 3]
print(f'b {b} id:{id(b)}')

b.append(4)

print(f'b {b} id:{id(b)}')
```
输出结果：
```python
a [1, 2, 3] id:140087109245896
b [1, 2, 3] id:140087109244168
b [1, 2, 3, 4] id:140087109244168
```

列表对象a和对象b虽然值一样，但是指向的内存地址不一样，因此就是可变的数据类型。列表内部增删元素并不会改变列表对象的内存地址！

### 2.1.1 python数字类型转换

python数据类型的转换，只要将数据类型最为函数名即可：

* int(x):将x转换为整数类型;
* float(x)：将x转换为浮点类型;
* complex(x)：将x转换为复数类型，实部为x，虚部为0;

### 2.1.2 十进制、二进制、八进制、十六进制数值类型转换

进制简述：

* `十进制`：就是最常见的数值类型，123
* `二进制`：0b1111011，数字0加字母b（binary）开头
* `八进制`：0o173，数字0加字母o（october）开头
* `十六进制`：0x7b，数字0加字母x（hex）开头

1、`十进制`转换为`其他进制`

* `十进制`转`二进制`：`bin(decimal)`返回二进制字符串
* `十进制`转`八进制`：`oct(decimal)`返回八进制字符串
* `十进制`转`十六进制`：`hex(decimal)`返回十六进制字符串


```python
>>> decimal = 1024
>>> binary = bin(decimal)
>>> binary
'0b10000000000'
>>> octal = oct(decimal)
>>> octal
'0o2000'
>>> hexadecimal = hex(decimal)
>>> hexadecimal
'0x400'
>>> 
```


2、`其他进制`转换为`十进制`

* `二进制`转`十进制`：int(binary_str, 2)
* `八进制`转`十进制`：int(octal_str, 2)
* `十六进制`转`十进制`：int(hexadecimal_str, 2)

```python
>>> binary = '0b10000000000'
>>> int(binary, 2)
1024
>>> octal = '0o2000'
>>> int(octal, 8)
1024
>>> hexadecimal = '0x400'
>>> int(hexadecimal, 16)
1024
>>> 

```

### 2.2.1 访问字符串中的值


1、字符串的索引有两种方式：

* `从前往后索引`：字符串索引是从`0`开始的，例如`Hello`的索引关系如下：
* `从后往前索引`：字符串从最后一个字符串开始索引为-1,然后一次往前

```python
0  1  2  3  4
H  e  l  l  o
-5 -4 -3 -2 -1

```


2、字符串的截取：字符串的截取是通过`[]`符号实现

使用第一种索引截取字符串

```python
>>> str = "Hello"
>>> str[0]
'H'
>>> str[0:3]
'Hel'
>>> str[0:]
'Hello'
>>> str[:3]
'Hel'
>>> 
```

使用第二种索引截取字符串

```python
>>> str[-1]
'o'
>>> str[:-1]
'Hell'
>>> str[-4:-2]
'el'
>>> str[-4:]
'ello'
>>> str[-5:-2]
'Hel'
>>> 
```

`注意：`

>无论是哪种索引截取都是`包前不包后`，例如：[2:5]截取的字符分别是索引2、3、4,并不包括后面的5

### 2.2.9 字符串的内建函数

注意：

以下所有的`S`都是值指代一个字符串对象S，不要误理解就是一个字母`S`


1、`S.capitalize()`：将字符串S的`首字母大写`

```python
>>> a = 'hello'
>>> a.capitalize()
'Hello'
>>> a
'hello'
```

2、`S.casefold()`：casefole()函数是python3.3之后版本引入的，其作用和`lower()`方法相似，都是将`字符串中所有的大写字母都转换为小写`


二者的区别：
* lower()方法只对`ASCII`编码有效，也即是对`A-Z`有效
* casefold()方法可以把`非汉语或英文`中的`大写转换为小写`

```python
>>> S1 = "Runoob EXAMPLE....WOW!!!" #英文
>>> S2 = "ß"  #德语
>>> print( S1.lower() )
runoob example....wow!!!
>>> print( S1.casefold() )
runoob example....wow!!!
>>> print( S2.lower() )
ß
>>> print( S2.casefold() ) #德语的"ß"正确的小写是"ss"
ss
>>> 
```

3、`S.center(width, fillchar)`：把字符串S填充到fillchar字符中间

* `width`:是填充字符和S字符串的总长度
* `fillchar`：是要在S字符串两边填充的字符

```python
>>> a = "Hi"
>>> a.center(10, "#")
'####Hi####'
```

4、`S.count(sub, start= 0,end=len(string))`：统计字符串里`某个字符`出现的次数。可选参数为在字符串搜索的开始与结束位置。

* sub：搜索的子字符串
* start：字符串开始搜索的位置。默认为第一个字符,第一个字符索引值为0。
* end：字符串中结束搜索的位置。字符中第一个字符的索引为 0。默认为字符串的最后一个位置。

```python
>>> str = "www.google.com"
>>> str.count("o")
3
>>> str.count("oo")
1
>>> str.count("ww")
1
>>> str.count("o", 10, 13)
1
```

5、`S.encode(encoding="utf-8", errors="strict")`：指定字符串的编码方式。如果出错默认是报`ValueError的异常`，除非errors指定的是`ignore`或`replace`

* `encoding`：是要使用的`编码方式`，如：`UTF-8`
* `errors`：置不同错误的处理方案。默认为 'strict',意为编码错误引起一个UnicodeError。 其他可能得值有 'ignore', 'replace', 'xmlcharrefreplace', 'backslashreplace' 以及通过 codecs.register_error() 注册的任何值。


可以使用：`chardet`查看字符的编码方式（在python3中已经不支持`S.decode()方法`）：

```python
>>> chardet.detect(b'abc123')
{'encoding': 'ascii', 'confidence': 1.0, 'language': ''}
```

`注意：`

>detect函数有一个参数, 即`非 unicode 字符串`。它`返回一个字典`, 其中包含自动检测的字符编码和从0到1的置信度。



6、`S.endswith(suffix[, start[, end]])`：用于判断字符串`是否以指定后缀结尾`，如果以`指定后缀结尾`返回`True`，否则返回`False`。可选参数 "start" 与 "end" 为检索字符串的开始与结束位置。


* `suffix`：该参数可以是一个字符串或者是一个元素;
* `start`：字符串中的开始位置;
* `end`：字符中结束位置。


```python
>>> str = "hello world!!"
>>> str.endswith("!!")
True
>>> str.endswith("!")
True
>>> str.endswith("!", 0, 5)
False
>>> str.endswith("o", 0, 5)  
True
>>> 
```

`注意：`

>start和end索引可以理解为先进行字符串的切片截取操作


7、`S.expandtabs(tabsize)`：改变转义符`\t`空格的个数，默认是空格一个tab长度，也就是四个字符。可以通过`tabsize`改变空格的长度。

```python
>>> str = 'how\tare\tyou?'
>>> print(str)
how	   are	 you?
>>> print(str.expandtabs(3))
how   are   you?
>>> print(str.expandtabs(1))
how are you?
```

8、`S.find(str, beg=0, end=len(string))`：查找字符串中是否包含`子字符串str`，如果包含返回`子字符串的索引`，否则返回`-1`。



* `str`：指定检索的字符串
* `beg`：开始索引，默认为0。
* `end`：结束索引，默认为字符串的长度。


```python
>>> str = "www.google.com"
>>> str.find("o")
5
>>> str.find("oo")  
5
>>> str.find("G")
-1
>>> 
```

`注意：`
>如果差查找的子字符串长度大于2,则返索引是子字符串第一个字符在查找字符串中匹配到字符串对应的第一个字符的索引。
